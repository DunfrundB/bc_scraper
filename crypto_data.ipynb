# Quick slicing of data from CoinAPI integration page to capture all currencies on all exchanges

```python
coinapi_tables = pd.read_html('https://www.coinapi.io/integration')
```

```python
# coinapi_tables contains all table objects from the url
# exhchanges and assets are used to capture the data from the tables list
exchanges = coinapi_tables[0]
assets = coinapi_tables[1]
```

```python
# assets are further divided into co_df and fo_df for Crypto Only and Fiat Only
co_df = assets[assets.Type == 'Crypto']
co_df.reset_index(drop=True,inplace=True)
co_df
```

```python
fo_df = assets[assets.Type != 'Crypto']
fo_df.reset_index(drop=True,inplace=True)
fo_df
```
